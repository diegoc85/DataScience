```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/diegoc85/Downloads/pokedex.csv")
df
```

```python
q1_total = df['Total'].quantile(0.25)
q3_total = df['Total'].quantile(0.75)
iqr_total = q3_total - q1_total
limite_inferior_total = q1_total - 1.5 * iqr_total
limite_superior_total = q3_total + 1.5 * iqr_total
total_out = df[(df['Total'] < limite_inferior_total) | (df['Total'] > limite_superior_total)]
total_outliers = total_out.shape[0]
print(f"Total de outliers encontrados: {total_outliers}")
              
```
