```python
# Exerc√≠cio 2 - Qual √© a m√©dia de for√ßa dos her√≥is da DC?

editora = df2[df2['Editora'] == 'DC']
media = editora['For√ßa'].mean()
media

```

```python
# Exercicio 3 - Qual editora tem a maior m√©dia de intelig√™ncia?

editora_DC = df2[df2['Editora'] == 'DC']
media_DC = editora_DC['Intelig√™ncia'].mean()

editora_MV = df2[df2['Editora'] == 'Marvel']
media_MV = editora_MV['Intelig√™ncia'].mean()

print(f"A m√©dia da intelig√™ncia da Marvel √© {media_MV} e da DC √© {media_DC}")

if media_DC > media_MV:
    maior = media_DC
    print(f"A editora com maior m√©dia √© a DC com {maior}.")
else:
    maior = media_MV
    print(f"A editora com maior m√©dia √© Marvel com {maior}. ")


```

```python
# Exercicio 4 - Qual √© o her√≥i mais veloz?

velocidade_max = df2['Velocidade'].max()
veloz = df2.loc[df2['Velocidade'] == velocidade_max]

print(veloz)




```

```python
# Exerc√≠cio 5 - Quantos her√≥is estrearam antes de 1960?
menores = df2[df2['Ano_Estreia'] < 1960]
quantidade = len(menores)
print(f"A quantidade de her√≥is que estreiaram antes de 1960 √© {quantidade}.")
```
