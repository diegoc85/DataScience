```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50) #qtd de números que irá gerar (números entre 0 e 1)
y = np.random.rand(50)
```

Exemplo 6: Heatmap / Mapa de Calor
- É uma matriz 

```python
import seaborn as sns
```

```python
%pip install seaborn
```

```python
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']
horas = ['Manhã', 'Tarde', 'Noite', 'Madrugada', 'Almoço', 'Happy Hour']
vendas = np.random.randint(0,50,(7,6)) # gera valores aleatórios de vendas
```

```python
#Criando o heatmap
plt.figure(figsize=(8,5))
sns.heatmap(vendas, annot=True, cmap= "coolwarm", xticklabels=horas, yticklabels=dias)
```

```python
plt.title("Heatmap de vendas por dia e período")
plt.xlabel("Período do dia ")
plt.ylabel("Dia da semana")
plt.show()
```

Exemplo 7 - Personalizando Gráficos 

```python
plt.plot(x,y, color = 'red', linestyle='--', marker='s', label='Série de Dados' )
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.grid(True)
plt.show()
```
