O código np.random.randn(1000) não gera valores aleatórios apenas entre 0 e 1 ou -1 e 1. Ele gera valores seguindo uma distribuição normal padrão, que tem: média = 0, desvio padrão = 1. Os números gerados não estão limitados a um intervalo fixo como [0,1] ou [-1,1]. Em vez disso, a maioria dos valores estará próxima de 0, mas alguns serão bem maiores ou menores. O bin representa está no eixo x, pois representa os intervalos de valores (é como falasse que tem 20 barras no gráfico). A altura da barra é a frequência em que ele apareceu. 

```python
import numpy as np
import matplotlib.pyplot as plt

dados = np.random.randn(1000) #media 0, Desvio padrão 1

plt.hist(dados, bins=30, color = 'blue', edgecolor='black', alpha = 0.7)
plt.axvline(x=0, color='red', linestyle='dashed', label='Média=0')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Distribuição Normal Padrão')
plt.legend()
plt.show()
```

```python
#DISTRIBUIÇÕES

#Controle de Qualidade - Distribuição normal
# Uma fábrica produz parafusos com comprimento médio de 100mm  e desvio padrão de 15mm.
# Qual a probabilidade de um parafuso ter mais de 120mm de comprimento? 

from scipy.stats import norm
mu = 100
sigma = 15
prob_normal = 1 - norm.cdf(120,mu, sigma)
print(f"Probabilidade de comprimento > 120mm: {prob_normal:.4f}")
```

```python
mu = 100
sigma = 15
prob_normal70 = norm.cdf(70,mu,sigma)
print(f"Probabilidade de comprimento de 0 a 70mm: {prob_normal:.4f}")

prob_normal80 = norm.cdf(80,mu,sigma)
print(f"Probabilidade de comprimento de 0 a 80mm: {prob_normal80:.4f}")

prob_normal_fim = prob_normal80 - prob_normal70
print(f"Probabilidad de comprimento de 70 a 80mm: {prob_normal_fim:.4f}")
```

```python
mu = 100
sigma = 15
prob_normal140 = norm.cdf(140,mu,sigma)
print(f"Probalidade de comprimento de 0 a 140mm: {prob_normal140}")

prob_normal_final = 1 - prob_normal140
print(f"Probabilidade maior que 140mm:{prob_normal_final:.4f}")
```

```python
# 90 mm a 95mm (11,7%)

mu = 100
sigma = 15
prob_normal90 = norm.cdf(90,mu,sigma)
print(f"Probabilidade de comprimento de 0 a 70mm: {prob_normal:.4f}")

prob_normal95 = norm.cdf(95,mu,sigma)
print(f"Probabilidade de comprimento de 0 a 80mm: {prob_normal80:.4f}")

prob_normal_finale = prob_normal95 - prob_normal90
print(f"Probabilidade de comprimento de 90 a 95mm: {prob_normal_finale:.4f}")
```

Você trabalha numa fábrica de tubos metáicos. O diâmetro dos tubos segue uma distribuição normal com: 
média(u) = 5cm 
desvio padrão (o) = 0.2 cm
Qual a probabilidade de um tubo ter diametro maior que 5,3 cm? 
Resposta: 6.7%

```python
mu = 5
sigma = 0.2
prob_normal_tubo_cinco = norm.cdf(5.3,mu,sigma)
print(f"Probabilidade de comprimento de 0 a 5.3: {prob_normal_tubo_cinco:.4f}")

prob_normal_maior_cinco = 1 - prob_normal_tubo_cinco
print(f"Probabilidade de comprimento maior que 5.3: {prob_normal_maior_cinco:.4f}")




```
